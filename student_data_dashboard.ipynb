```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create the dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'Math': [78, 92, 65, 88, 73],
    'English': [85, 81, 74, 90, 69],
    'Science': [88, 79, 70, 85, 75]
}
df = pd.DataFrame(data)

# Step 3: Show dataset
print("Student Scores:\n", df)

# Step 4: Averages, Min, Max
print("\nAverage per subject:\n", df.mean(numeric_only=True))
print("\nMaximum scores:\n", df.max(numeric_only=True))
print("\nMinimum scores:\n", df.min(numeric_only=True))

# Step 5: Bar Chart - Student Scores
df.plot(x='Name', kind='bar', title='Student Scores')
plt.tight_layout()
plt.savefig("student_scores_bar_chart.png")  # Saves in the same folder
plt.show()

# Step 6: Pie Chart - Total per Subject
subject_totals = df[['Math', 'English', 'Science']].sum()
plt.pie(subject_totals, labels=subject_totals.index, autopct='%1.1f%%')
plt.title('Subject-wise Total Performance')
plt.tight_layout()
plt.savefig("subject_pie_chart.png")  # Saves in the same folder
plt.show()

# Step 7: Save to Excel file
with pd.ExcelWriter("student_data_dashboard.xlsx") as writer:
    df.to_excel(writer, sheet_name='Student Scores', index=False)
    df.mean(numeric_only=True).to_frame(name='Average').to_excel(writer, sheet_name='Averages')
    df.max(numeric_only=True).to_frame(name='Maximum').to_excel(writer, sheet_name='Maximums')
    df.min(numeric_only=True).to_frame(name='Minimum').to_excel(writer, sheet_name='Minimums')
```
